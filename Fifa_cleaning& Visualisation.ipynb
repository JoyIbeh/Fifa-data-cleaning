```python
def clean_fifa_data(df):
    """
    Cleans the 'Value', 'Wage', and 'Release Clause' columns in the FIFA dataset by removing currency symbols 
    and converting letters to numerical representations.
    
    Args:
    df (pd.DataFrame): The FIFA dataset containing the columns 'Value', 'Wage', and 'Release Clause'.
    
    Returns:
    pd.DataFrame: The cleaned FIFA dataset.
    """
    
    # Define a dictionary for replacements
    replacements = {
        '€': '',
        'M': 'e6',
        'm': 'e6',
        'K': 'e3',
        'k': 'e3'
    }
    
    # Apply replacements to 'Value' column
    for old, new in replacements.items():
        df['Value'] = df['Value'].str.replace(old, new)
    
    # Apply replacements to 'Wage' column
    for old, new in replacements.items():
        df['Wage'] = df['Wage'].str.replace(old, new)
    
    # Apply replacements to 'Release Clause' column
    for old, new in replacements.items():
        df['Release Clause'] = df['Release Clause'].str.replace(old, new)
    
    return df

# Example usage:
# fifa_data = clean_fifa_data(fifa_data)
```

```python
def clean_fifa_data(df):
    """
    Cleans the 'Value', 'Wage', and 'Release Clause' columns in the FIFA dataset by removing currency symbols 
    and converting letter representations (M, K) to numerical representations.
    
    Args:
    df (pd.DataFrame): The FIFA dataset containing the columns 'Value', 'Wage', and 'Release Clause'.
    
    Returns:
    pd.DataFrame: The cleaned FIFA dataset with numeric 'Value', 'Wage', and 'Release Clause'.
    """
    
    # Define a function to clean a single column
    def clean_column(column):
        column = column.str.replace('€', '', regex=False)   # Remove the Euro symbol
        column = column.str.replace('M', '000000', regex=False)  # Convert 'M' to '000000'
        column = column.str.replace('m', '000000', regex=False)  # Convert 'm' to '000000'
        column = column.str.replace('K', '000', regex=False)     # Convert 'K' to '000'
        column = column.str.replace('k', '000', regex=False)     # Convert 'k' to '000'
        column = column.str.replace('.', '', regex=False)        # Remove any remaining periods
        return pd.to_numeric(column, errors='coerce')  # Convert to numeric (float) and handle errors
    
    # Apply the cleaning function to each column
    df['Value'] = clean_column(df['Value'])
    df['Wage'] = clean_column(df['Wage'])
    df['Release Clause'] = clean_column(df['Release Clause'])
    
    return df

# Example usage:
fifa_data = clean_fifa_data(fifa_data)
```

```python
def clean_star_columns(df, columns):
    """
    Removes star characters from the specified columns in a DataFrame and converts them to integers.

    Parameters:
    df (pd.DataFrame): The DataFrame containing the columns to clean.
    columns (list of str): List of column names to clean in the DataFrame.

    Returns:
    pd.DataFrame: The DataFrame with cleaned columns.
    """
    for column in columns:
        df[column] = df[column].str.replace('★', '').astype(int)
    return df

# Usage example:
fifa_data = clean_star_columns(fifa_data, ['IR', 'W/F', 'SM'])
```
